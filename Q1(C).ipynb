```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load both grayscale images
img1 = cv2.imread("image1.jpeg", 0)
img2 = cv2.imread("image2.jpeg", 0)

# Check if images loaded
if img1 is None:
    raise ValueError("image1.jpeg not found. Check file name or path.")
if img2 is None:
    raise ValueError("image2.jpeg not found. Check file name or path.")

def gamma_correction(image, gamma):
    # Convert to float and normalize (important for accuracy)
    normalized = image.astype(np.float32) / 255.0
    
    # Apply gamma transformation
    corrected = np.power(normalized, gamma)
    
    # Scale back to 0-255
    corrected = np.uint8(corrected * 255)
    
    return corrected

# Gamma values
gamma_values = [0.4, 0.67, 1.0, 1.5, 2.5]

def show_gamma_results(img, title):
    plt.figure(figsize=(15,5))
    
    # Original Image
    plt.subplot(1,6,1)
    plt.imshow(img, cmap='gray')
    plt.title("Original")
    plt.axis('off')
    
    # Gamma Outputs
    for i, gamma in enumerate(gamma_values):
        corrected_img = gamma_correction(img, gamma)
        
        plt.subplot(1,6,i+2)
        plt.imshow(corrected_img, cmap='gray')
        plt.title(f"γ = {gamma}")
        plt.axis('off')
    
    plt.suptitle(title)
    plt.tight_layout()
    plt.show()

# Display results for both images
show_gamma_results(img1, "Gamma Transformation - Image 1")
show_gamma_results(img2, "Gamma Transformation - Image 2")
```
